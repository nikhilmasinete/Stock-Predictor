```python
import pandas as pd
import datetime
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import pandas_datareader as dr
from sklearn.model_selection import train_test_split


```

The data of a particular company's stock is imported from the Yahoo Finance.
The start date andthe end date of the stocks must be prescribed.

```python
cmp_name = "GOOG"
start = datetime.datetime(2010,1,1)
end = datetime.datetime(2017,1,11)
cmp_stks = dr.get_data_yahoo(cmp_name,start,end)
cmp_stks.info()
```

```python
cmp_stks['Adj Close'].plot()
```

# Data Preparation
The data is prepared for training and test set.
The parameters Adj Close and Close are almost the same.
The input matrix consists of the High, Low, Open and Volume.
The model should predict the Closing Value.

```python
y = cmp_stks['Close']
X = cmp_stks.drop(['Close','Adj Close'], axis = 1)
```

```python
mean_y = y.mean()
std_y = y.std()
new_y = (y - mean_y)/std_y
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(X)
scaled_x = scaler.transform(X)
X.info()
```

```python
X_train, X_val = train_test_split(scaled_x, test_size = 0.05, shuffle = False)
y_train, y_val = train_test_split(new_y, test_size = 0.05, shuffle = False)
```

```python

X_tr = np.array(X_train)
X_vala = np.array(X_val)
y_tr = np.array(y_train)
y_vala = np.array(y_val)

```

# Modelling

Here in exercise we use The following models to train our data.

1.Linear Regression
2.Lasso Regression
3.Support Vector Machine
4.Elastic nets
5.SGD Regressor


```python
from sklearn.linear_model import LinearRegression
linear = LinearRegression()
linear.fit(X_train, y_train)
linear_predict = linear.predict(X_val)
linear_score = linear.score(X_val, y_val)
linear_score
```

```python
df_pred_linear = pd.DataFrame({'pred':linear_predict})
df_tot_y = pd.concat([y_val, df_pred_linear], axis=1)
df_tot_y.info()
```

```python
plt.plot(linear_predict,'r', label = 'Predicted')
plt.plot(y_vala, 'b', label = 'Real')
plt.legend(loc = 'upper right')
plt.title('Linear')
```

The X axis denotes the date column. The model is predicting good values.

```python
from sklearn.linear_model import SGDRegressor

sgd_regressor = SGDRegressor()
sgd_regressor.fit(X_train,y_train)
sgd_predict = sgd_regressor.predict(X_val)

plt.plot(sgd_predict, 'r', label = 'Predicted')
plt.plot(y_vala,'b', label='real')
plt.legend(loc='upper right')
plt.title('SGD')
```

```python
from sklearn.neighbors import KNeighborsRegressor
knn = KNeighborsRegressor()
knn.fit(X_tr,y_tr)
knn_predict = knn.predict(X_vala)
knn_score = knn.score(X_vala,y_vala)
knn_score
```

```python
plt.plot(knn_predict,'r', label = 'predicted')
plt.plot(y_vala,'b', label = 'real')
plt.legend(loc='upper right')
plt.title('KNN_scaled')
```

```python
from sklearn.svm import SVR
svm_rbf = SVR(kernel='rbf')
svm_linear = SVR(kernel = 'linear')
svm_poly = SVR(kernel = 'poly')
svm_rbf.fit(X_tr,y_tr)
svm_linear.fit(X_tr,y_tr)
svm_poly.fit(X_tr,y_tr)
svm_rbf_predict = svm_rbf.predict(X_vala)
svm_linear_predict = svm_linear.predict(X_vala)
svm_poly_predict = svm_poly.predict(X_vala)
plt.plot(svm_rbf_predict, 'r+', label = 'SVM rbf')
plt.plot(svm_linear_predict, 'b', label = 'SVM lp')
plt.plot(y_vala, label = 'real')
plt.legend(loc = 'upper right')
```

```python
from sklearn.linear_model import ElasticNet
elastic_net = ElasticNet()
elastic_net.fit(X_tr,y_tr)
elastic_net_predict = elastic_net.predict(X_vala)
plt.plot(elastic_net_predict,'r', label='predicted')
plt.plot(y_vala,'b',label='real')
plt.legend(loc = 'upper right')
plt.title('ElasticNet')
```

# Conclusion
As a conclusion it can be said that the Linear regression and SVM linear Predict are good models for this data set.
